# Draw a Bar Plot using Matplotlib

Write a Python program to demonstrate how to draw a bar plot using Matplotlib.

## Problem Statement and Introduction

"In this program, we will create a bar plot using Matplotlib to visualize the number of books read by a group of students in each month."



"Let's write Python code to create a bar plot that represents the monthly books read by students. We have data for five months (Jan, Feb, Mar, Apr, and May) and the number of books read in each month."

````python
%matplotlib inline
````

## Step 1: Import Matplotlib

Import the necessary libraries for plotting:

````python
# Step 1: Step 1: Import Matplotlib
import matplotlib.pyplot as plt
````

## Step 2: Prepare Data

Define the data for months and books_read:

````python
# Step 2: Prepare Data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
books_read = [10, 12, 15, 9, 11]
````

## Step 3: Create a Bar Plot

Write Python code to create the bar plot:

````python
# Step 3: Create a Bar Plot
plt.bar(months, books_read, color='skyblue')
````

## Step 4: Customize the Plot

````python
# Step 4: Customize the Plot
plt.bar(months, books_read, color='skyblue')
plt.xlabel('Months')
plt.ylabel('Books Read')
plt.title('Monthly Books Read by Students')
````

##Step 5: Display the Plot


````python
#Step 5: Display the Plot
plt.bar(months, books_read, color='skyblue')
plt.xlabel('Months')
plt.ylabel('Books Read')
plt.title('Monthly Books Read by Students')
plt.show()
````

1. Question: Create a bar plot to visualize the sales of two products (Product A and Product B) in a year.

You have the following data:

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

product_a_sales = [2500, 2800, 3100, 3200, 3500, 3800, 4000, 4200, 4500, 4800, 5000, 5300]

product_b_sales = [2000, 2200, 2400, 2600, 2800, 3000, 3200, 3400, 3600, 3800, 4000, 4200]


````python
import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
product_a_sales = [2500, 2800, 3100, 3200, 3500, 3800, 4000, 4200, 4500, 4800, 5000, 5300]
product_b_sales = [2000, 2200, 2400, 2600, 2800, 3000, 3200, 3400, 3600, 3800, 4000, 4200]

# Create a bar plot for Product A
plt.bar(months, product_a_sales, label='Product A', color='skyblue')

# Create a bar plot for Product B on top of Product A
plt.bar(months, product_b_sales, label='Product B', color='lightgreen', alpha=0.7)

# Adding labels and a title
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Sales of Products A and B in a Year')

# Adding a legend
plt.legend()

# Adjusting the x-axis tick labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()  # To ensure all elements fit in the figure
plt.show()
````

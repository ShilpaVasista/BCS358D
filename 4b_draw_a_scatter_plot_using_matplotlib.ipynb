## Problem Statement and Introduction
1. Objective: Learn how to create a  Scatter plot using Matplotlib.

Imagine a scenario where students want to analyze the relationship between the number of hours they spend studying and their test scores. This scatter plot can help them understand if there's a correlation between study hours and test performance.

Use the provided data for study hours and test scores, represented as lists.
Create a scatter plot to visualize the relationship between study hours and test scores.

Customize the plot with appropriate labels for the x-axis and y-axis and give it a title.

Analyze the scatter plot to identify any trends or correlations between study hours and test scores.

Data
```Python

# Study hours (in hours)
study_hours = [2, 3, 4, 5, 6, 7, 8, 9, 10]

# Test scores (out of 100)
test_scores = [60, 65, 70, 75, 80, 85, 90, 95, 100]

```

````python
%matplotlib inline
````

## Step 1: Import Matplotlib

````python
# Step 1: Import Matplotlib
import matplotlib.pyplot as plt
````

## Step 2: Prepare Data
Sample data for student study hours and test scores
```
study_hours = [2, 3, 4, 5, 6, 7, 8, 9, 10]
test_scores = [60, 65, 70, 75, 80, 85, 90, 95, 100]
```

````python
# Step 2: Prepare Data
study_hours = [2, 3, 4, 5, 6, 7, 8, 9, 10]
test_scores = [60, 65, 70, 75, 80, 85, 90, 95, 100]
````

## Step 3: Create a Scatter Plot

````python
# Step 3: Create a Scatter Plot
plt.scatter(study_hours, test_scores, color='blue', marker='o', label='Student Data')
````

## Step 4: Customize the Plot

````python
# Step 4: Customize the Plot

plt.scatter(study_hours, test_scores, color='blue', marker='o', label='Student Data')
plt.xlabel('Study Hours')
plt.ylabel('Test Scores')
plt.title('Study Hours vs. Test Scores')
plt.legend()
````

##Step 5: Display the Plot

````python
#Step 5: Display the Plot

plt.scatter(study_hours, test_scores, color='blue', marker='o', label='Student Data')
plt.xlabel('Study Hours')
plt.ylabel('Test Scores')
plt.title('Study Hours vs. Test Scores')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
````
